```python
#Importamos librer√≠as requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
```

```python
#convertir archivo csv en dataframe 
data=pd.read_csv('lastmile_delivery_operations_mit_lift_lab.csv')
data
```

```python
#obtener informaci√≥n del dataframe
data.info()
```

```python
#corroboramos valores nulos
datosnulos=data.isnull().sum()
datosnulos
```

```python
#Separamos valores cuantitativos de valores cualitativos
filtro1=data.iloc[ : , [5,6,12,21,22,25,28]]
filtro1
```

```python
#Encontramos valores nulos por columna
nulos=filtro1.isnull().sum()
nulos
```

```python
#hacer una copia del dataframe
data2=filtro1.copy()
```

```python
#Reemplazar valores nulos con la funci√≥n promedio
data2l=data2.fillna(round(data2.mean()))
data2l
```

```python
#Eliminamos columa 'describe_the_accident', pues est√° totalmente vac√≠a
#Eliminamos columnas innecesarias, ya que estas columnas est√°n llenas en su totalidad de valores nulos
data2l2= data2l.drop(['describe_the_accident'], axis=1)
data2l2
```

```python
#corroboramos valores nulos
valores_nulos=data2l2.isnull().sum()

valores_nulos
```

```python
#M√©todo aplicando cuartiles para todo el dataframe
y=data2l2
Limite_Superior= y.mean() +3*y.std()
Limite_Inferior= y.mean() -3*y.std()
print("Limite superior permitido", Limite_Superior)
print("L√≠mite inferior permitido", Limite_Inferior)
```

```python
#Encontramos outliers del dataframe
outliers= data2l2[(y>Limite_Superior)|(y>Limite_Inferior)]
outliers
```

```python
#obtenemos datos sin outliers del dataframe
data3=data2l2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#corroboramos valores nulos del dataframe 
framelimpio=data3.isnull().sum()
framelimpio
```

```python
#Reemplazar valores nulos con promedio 
framesinnulos=data3.fillna(round(data3.mean(),1))
framesinnulos
```

```python
#corroboramos valores nulos del dataframe 
framelimpio1=framesinnulos.isnull().sum()
framelimpio1
```

```python
#Aplicar Filtro por columnas para encontrar valores cualitativos
columnacualitativa=data.iloc[ : , [0,1,2,3,4,7,8,9,10,11,13,14,15,16,17,18,19,20,23,24,26,27]]
columnacualitativa.info()
```

```python
#Buscar valores nulos en columna cualitativa
nuloscolumnacualitativa=columnacualitativa.isnull().sum()

nuloscolumnacualitativa
```

```python
#Limpiamos valores nulos de las columnas cualitativas 
columnacualitativa2= columnacualitativa.fillna('no encontrado')
columnacualitativa2
```

```python
#Comprobamos nulos
columnacualitativa3=columnacualitativa2.isnull().sum(1)

columnacualitativa3
```

```python
#unimos columnas cualitativas y cuantitativas
Dataframefinal=pd.concat([framesinnulos,columnacualitativa2])
```

```python
#convertir dataframe a excel
Dataframefinal.to_excel("lastmile_delivery_operations_mit_lift_lab.xlsx")
```
